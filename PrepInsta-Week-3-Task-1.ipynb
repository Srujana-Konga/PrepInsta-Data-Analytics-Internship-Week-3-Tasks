## Question 3

```python
def find_outliers(column_name):
    q1 = column_name.quantile(0.25)
    q3 = column_name.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = (column_name < lower_bound) | (column_name > upper_bound)
    return outliers
```

```python
find_outliers(df['Age'])
```

```python
find_outliers(df['Salary'])
```
